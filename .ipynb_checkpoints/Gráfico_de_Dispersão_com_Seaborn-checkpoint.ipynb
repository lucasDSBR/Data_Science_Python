```python
import pandas as pd
import seaborn as srn
import matplotlib.pyplot as plt
```

```python
base = pd.read_csv('./dados/co2.csv')
base.head()
```

```python
#Gráficos de dispersão utilizando os atributos conc e update, agrupamento pelo type
srn.scatterplot(base.conc, base.uptake, hue = base.Type)
```

```python
#Seleção de registros especificos da base de dados (Quebec e Mississipi)
q = base.loc[base['Type'] == 'Quebec']
m = base.loc[base['Type'] == 'Mississippi']
```

```python
#Subgráfico(1 linha e duas colunas) mostrando gráficos sobre cada registro
plt.figure()
plt.subplot(1, 2, 1)
srn.scatterplot(q.conc, q.uptake).set_title('Quebec')
plt.subplot(1, 2, 2)
srn.scatterplot(m.conc, m.uptake).set_title('Mississippi')
plt.tight_layout()
```

```python
#Refrigrado e não refrigerado
ch = base.loc[base['Treatment'] == 'chilled']
nc = base.loc[base['Treatment'] == 'nonchilled']
```

```python
#Gráficos somente com 'chilled' e 'nonchilled'
plt.figure()
plt.subplot(1, 2, 1)
srn.scatterplot(ch.conc, ch.uptake).set_title('Chilled')
plt.subplot(1, 2, 2)
srn.scatterplot(nc.conc, nc.uptake).set_title('Non Chilled')
plt.tight_layout()
```

```python
#Carregando dados de outro arquivo
base2 = pd.read_csv('./dados/esoph.csv')
base2.head()
```

```python
#Gráfico entre os atributos 'alcgp' e 'ncontrols'
srn.catplot(x = 'alcgp', y = 'ncontrols', data = base2, jitter = False)
```

```python
#Gráfico entre os atributos 'alcgp' e 'ncontrols', com agrupamento
srn.catplot(x = 'alcgp', y = 'ncontrols', data = base2, col = 'tobgp')
```
