```python
#Filtro
dados.loc[dados['purpose'] == "radio/tv"]
```

```python
#Atribuindo valores a outra variavel, criando outro df
credito2 = dados.loc[dados['credit_amount'] > 18000]
print(credito2)
```

```python
#Definindo só algunmas colunas
credito3 = dados[['checking_status', 'duration']].loc[dados['credit_amount'] > 10000]
print(credito3)
```

```python
#Série, única coluna
# Pode ser criada a partir de listas, array do numpy ou coluna de data frame
s1 = pd.Series([2, 5, 3, 34, 54, 23, 1, 16])
print(s1)
```

```python
# Series a partir de um array do numpy
import numpy as np
array1 = np.array([2, 5, 3, 34, 54, 23, 1, 16])
s2 = pd.Series(array1)
print(s2)
```

```python
#Série a partir de um data frame
s3 = dados['purpose']
print(s3)
```
